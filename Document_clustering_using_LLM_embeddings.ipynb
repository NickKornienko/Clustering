```python
from transformers import AutoTokenizer, AutoModel
from sklearn.cluster import KMeans
import torch
import numpy as np
```

```python
# Load BERT LLM
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
model = AutoModel.from_pretrained("bert-base-uncased")
```

```python
# Transform text data to embeddings
def get_embedding(text):
    # Ensure that the input is a string
    if not isinstance(text, str):
        raise ValueError("Input text must be a string.")

    # Tokenize and get embeddings
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=512)
    outputs = model(**inputs)
    return outputs.last_hidden_state.mean(axis=1).detach().numpy()
```

```python
# Load document data (synthetic)
technology_texts = [
    "Tech Article 1: Exploring the advancements in artificial intelligence and machine learning.",
    "Tech Article 2: The impact of blockchain technology on financial industries.",
    "Tech Article 3: Future of quantum computing and its potential applications.",
    "Tech Article 4: The rise of Internet of Things (IoT) in smart home systems.",
    "Tech Article 5: Autonomous vehicles and the revolution in transportation.",
]

sports_texts = [
    "Sports News 1: Latest updates from the world of soccer.",
    "Sports News 2: The rise of basketball as a global sport.",
    "Sports News 3: Track and field events break new records.",
    "Sports News 4: The changing world of competitive swimming.",
    "Sports News 5: Tennis stars set new benchmarks.",
]

cooking_texts = [
    "Cooking Recipe 1: How to bake a classic chocolate cake.",
    "Cooking Recipe 2: The art of making perfect pasta.",
    "Cooking Recipe 3: Secrets of a delicious homemade pizza.",
    "Cooking Recipe 4: Exploring the flavors of Indian cuisine.",
    "Cooking Recipe 5: Mastering the barbecue grill.",
]

travel_texts = [
    "Travel Guide 1: Exploring the beauty of the Italian countryside.",
    "Travel Guide 2: A journey through the historic cities of Europe.",
    "Travel Guide 3: Discovering the hidden gems of Southeast Asia.",
    "Travel Guide 4: The ultimate guide to backpacking in South America.",
    "Travel Guide 5: Luxury travel destinations for your next vacation.",
]

literature_texts = [
    "Literature Review 1: Analysis of modernist themes in early 20th-century literature.",
    "Literature Review 2: The evolution of the novel form over centuries.",
    "Literature Review 3: Exploring the works of Shakespeare in contemporary times.",
    "Literature Review 4: The impact of digital media on traditional storytelling.",
    "Literature Review 5: Postcolonial narratives and their significance.",
]

import random

# Combine all texts into one list
all_texts = technology_texts + sports_texts + cooking_texts + travel_texts + literature_texts

# Shuffle the combined list
random.shuffle(all_texts)
```

```python
# generate embeddings
doc_embeddings = np.vstack([get_embedding(doc) for doc in all_texts])
```

```python
# Apply clustering
n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters)
kmeans.fit(doc_embeddings)
labels = kmeans.labels_
```

```python
# Visualize results
for i in range(n_clusters):
    print(f"\nCluster {i}:")
    for doc, label in zip(all_texts, labels):
        if label == i:
            print(f" - {doc}")
```
