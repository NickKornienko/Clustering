```python
import numpy as np
import pandas as pd
from pyod.models.knn import KNN
from pyod.utils.data import generate_data
import matplotlib.pyplot as plt
```

```python
# Synthetic data
X_train, X_test, y_train, y_test = generate_data(n_train=200, n_test=100, n_features=2, contamination=0.1, random_state=42)
```

```python
# Create and Train the Anomaly Detection Model
clf = KNN()
clf.fit(X_train)
```

```python
# Predict anamolies
y_test_pred = clf.predict(X_test)
```

```python
# Evalutate the model
n_errors = (y_test_pred != y_test).sum()
print(f'Number of errors: {n_errors}')
```

```python
# Visualize the results
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test_pred)
plt.title("Anomaly Detection")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
